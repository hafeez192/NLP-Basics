```python
import nltk
```

```python
from nltk.tokenize import word_tokenize
text = 'Ai is the coined in 1956 but it gained popularity in 2012'
print(word_tokenize(text))
```

```python
from nltk.tokenize import sent_tokenize
text = 'Ai is the coined in 1956 but it gained popularity in 2012. This is amazing thing'
print(sent_tokenize(text))
```

```python
from nltk.stem import PorterStemmer
e_words = ['wait','waited','waiting','waits']
ps = PorterStemmer()
for w in e_words:
    rootWord = ps.stem(w)
    print(rootWord)
```

```python
from nltk.stem.porter import PorterStemmer
porter_stemmer = PorterStemmer()
text ='Studies studying cries cry'
tokenization = nltk.word_tokenize(text)
for w in tokenization:
    print(f'stemming for {w} is {porter_stemmer.stem(w)}')
```

```python
from nltk.stem import WordNetLemmatizer
wordnet_lemma = WordNetLemmatizer()
text ='Studies studying cries cry'
tokenization = nltk.word_tokenize(text)
for w in tokenization:
    print(f'Lemma for {w} is {wordnet_lemma.lemmatize(w)}')
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
data = 'AI was introduced in the year 1956 but gained popularity recently'
stopWords = set(stopwords.words('english'))
words = word_tokenize(data)
wordsFiltered = []

for w in words:
    if w not in stopWords:
        wordsFiltered.append(w)

print(wordsFiltered)
```
